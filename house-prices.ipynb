```python
data = df_train.select_dtypes(include=[np.number]).interpolate().dropna()
data.isnull().sum()
```

```python
# Construindo o modelo
y = np.log(data['SalePrice'])
X = data.drop(['SalePrice', 'Id'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
                          X, y, random_state=42, test_size=.33)
```

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
```

```python
model = lr.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

Criando submissão para o Kaggle

```python
submission = pd.DataFrame()
submission['Id'] = df_test['Id']

features = df_test.select_dtypes(
        include=[np.number]).drop(['Id'], axis=1).interpolate()

predictions = model.predict(features)

result = np.exp(predictions)
submission['SalePrice'] = result

submission.head(3)
```

```python
submission.to_csv('./house_prices_linear_regression.csv', index=False)
```
